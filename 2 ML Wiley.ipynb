```python
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
sent_tokens = nltk.sent_tokenize(textdata)
print("Sentence tokens are:\n", sent_tokens)

#tokenize document
doc_tokens = nltk.word_tokenize(textdoc)
print("The word tokens are\n", doc_tokens)
```

```python
# removal of stopwords
nltk.download('stopwords')

eng_stopwords = nltk.corpus.stopwords.words('english')

print(len(eng_stopwords))

print(eng_stopwords[0:30])
```

```python
#filtering

final_tokens = [ token for token in doc_tokens if token not in eng_stopwords]

newdoc = ' '.join(final_tokens)
print(len(newdoc))

print(newdoc)
```

```python
# creating bag of words

from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
import pandas as pd
```

```python
print('Number of senetences are:', len(sent_tokens))
```

```python
# remove unwanted elements from data like symbols, numbers
token = RegexpTokenizer(r'[a-zA-Z0-9]+')

#count vectorizer to display the words
cv = CountVectorizer(lowercase=True, stop_words='english', ngram_range = (1,1), tokenizer = token.tokenize)
text_counts = cv.fit_transform(sent_tokens)

#creating info of words
print("information of words:\n", text_counts.toarray())
```

```python
#creating doc term matrix
count_df = pd.DataFrame(text_counts.toarray(), columns = cv.get_feature_names_out())

#newer version has function cv.get_feature_names()

#display the shape of datagrame
print("Shape of dataframe", count_df.shape)

#display the bag of words
print(count_df)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

nltk.download('wordnet')


ps = PorterStemmer()
lem = WordNetLemmatizer()

mywords = ["program", "programs", "programming", "programmer", "programmers"]
for word in mywords:
    print("Stemmer:", word, ": ", ps.stem(word))
    print("Leema ", word, ": ", lem.lemmatize(word,'v'))
```

```python
# stemmer and lemma on txt doc

lemlist = []
pslist = []

for word in final_tokens:
    lemlist.append(lem.lemmatize(word,'v'))
    pslist.append(ps.stem(word))
    
print("After lemmization: \n", lemlist[0:30])
print("After stemming:\n", pslist[0:30])
```

# Working with spaCy (adavanced natural language library)

```python
import pandas as pd

data = pd.read_csv("mcdonalds.csv")
data.head()
```

```python
alldata =''
for i in range(0, len(data)):
    alldata= alldata+data['review'][i]
    
print(len(alldata))
```

```python
#Text preprocessing of the column
def preprocess(column_name):
    column_name = column_name.str.lower()
    column_name = column_name.str.strip()
    column_name = column_name.str.replace(r"^[a-zA-Z]\s+", "")
    column_name = column_name.str.replace(r"","")
    column_name = column_name.str.replace(r"#","")
    column_name = column_name.str.replace(r"wouldn't","would not")
    column_name = column_name.str.replace(r"mcdonald' s","mcdonald")
    column_name = column_name.str.replace(r"mcdonald's","mcdonald")
    column_name = column_name.str.replace(r"mcdonalds","mcdonald")
    column_name = column_name.str.replace(r"i\'m","i am")
    column_name = column_name.str.replace(r"it\'s","it is")
    column_name = column_name.str.replace(r"i\'ve","i have")
    column_name = column_name.str.replace(r"ain't","am not")
    column_name = column_name.str.replace(r"isn't","is not")
    column_name = column_name.str.replace(r"aren't","are not")
    column_name = column_name.str.replace(r"can't","can not")
    column_name = column_name.str.replace(r"shouldn't","should not")
    column_name = column_name.str.replace(r"shan't","shall not")
    column_name = column_name.str.replace(r"won't","will not")
    column_name = column_name.str.replace(r"hasn't","has not")
    column_name = column_name.str.replace(r"hadn't","had not")
    column_name = column_name.str.replace(r"don't","do not")
    column_name = column_name.str.replace(r"couldn't","could not")
    column_name = column_name.str.replace(r"didn't","did not")
    column_name = column_name.str.replace(r"aint","am not")
    column_name = column_name.str.replace(r"isnt","is not")
    column_name = column_name.str.replace(r"wouldnt","would not")
    column_name = column_name.str.replace(r"arent","are not")
    column_name = column_name.str.replace(r"cant","can not")
    column_name = column_name.str.replace(r"shouldnt","should not")
    column_name = column_name.str.replace(r"shant","shall not")
    column_name = column_name.str.replace(r"wont","will not")
    column_name = column_name.str.replace(r"hasnt","has not")
    column_name = column_name.str.replace(r"hadnt","had not")
    column_name = column_name.str.replace(r"dont","do not")
    column_name = column_name.str.replace(r"couldnt","could not")
    column_name = column_name.str.replace(r"didnt","did not")
    column_name = column_name.str.replace(r"didnt","did not")
    return column_name

data['ProcessReview'] = preprocess(data['review'])
textdata=""
for i in range(0,len(data)):
    textdata = textdata+data['ProcessReview'][i]

print(textdata[0:20])
print("Numbers of letter after preprocessing", len(textdata))
```

```python
# Named entity recognition

import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(textdata)

#displaying entities and labe
print("Entities and labels")
for entity in doc.ents:
    print(entity, '=>', entity.label_)
```

```python
from collections import Counter
entity_data = [entity.text for entity in doc.ents]
print("Most commont words\n",Counter(entity_data).most_common(20))
```

```python
#tokenization

sentence_count = 0
for sent in doc.sents:
    sentence_count+=1

print("No of sentences in docu", sentence_count)
```

```python
#creating list of tokens
spacy_token = []
for token in doc:
    spacy_token.append(token)
print("Number of tokens are:", len(spacy_token))
print("tokens are\n", spacy_token[0:20])
```

```python
#Stop word removal
from spacy.lang.en.stop_words import STOP_WORDS

final_token = [token for token in doc if not token.is_stop and not token.is_punct and not token.is_digit]

print('Number of tokens after removal of stop words', len(final_token))
print('tokes are now:--', final_token[0:30])
```

```python
#shallow parsing: pos tagging
i=0
for token in final_token:
    print(token.text, "==>", token.pos_)
    i+=1
    if i > 20:
        break
```

```python
#determining total number of words for different parts of speech
nc, vc, pn, ac, advc=0,0,0,0,0
for token in final_token:
    if token.pos_ == 'NOUN':
        nc+=1
    if token.pos_ =='VERB':
        vc+=1
    if token.pos_ == 'PROPN':
        pn+=1
    if token.pos_ =='ADJ':
        ac+=1
    if token.pos_ == 'ADV':
        advc+=1

pos_dict = {'Noun': nc, 'Verb': vc, 'Propostion': pn, 'Adjective': ac, 'Adverb': advc}
print(pos_dict)
```

```python
i=0
for token in final_token:
    print(token.text, "==>", token.tag_)
    i+=1
    if i > 20:
        break
```

```python
#lemmatiztion
lemma_token =  []

for token in final_token:
    lemma_token.append(token.lemma_)

print(lemma_token[0:30])
```

```python
# frequency distributions

from collections import defaultdict
frequency = defaultdict(int)

for token in lemma_token:
    frequency[token]+=1

print("Dictionary of word and no of occ \n")
i=0
for i,(key,value) in enumerate(frequency.items()):
    if i> 15:
        break
    print(f"{key}:{value}")
```

```python
#creating dataframe from dictionary
mydf = pd.DataFrame([frequency]).T
mydf['Word'] = mydf.index
mydf.reset_index(drop =True, inplace =True)
mydf = mydf.rename(columns = {0:'Count'}, inplace = False)
print(mydf.head(5))
```

```python
print(mydf.sort_values(by='Count', ascending =False).head(10))
```
